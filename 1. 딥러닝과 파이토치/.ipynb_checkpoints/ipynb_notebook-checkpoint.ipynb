# 1.1. 인공지능과 머신러닝

## 1.1.2. 초기 인공지능 VS 머신러닝

* 초기 인공지능
: 초기 인공지능 개발시에 학습을 위해서, '자동차'라는 오브젝트를 판별하기 위해, 자동차의 모든 특성(바퀴 수, 창문 등)을 인지시켜주어야 했다.

* 머신러닝
: 초기 인공지능 개발과 달리 단지 입력값에 대한 결과값만 주어주면, 결과 값 예측을 위해 스스로 자동차의 특성을 학습하는 것.

# 1.2. 지도학습, 비지도학습, 강화학습

* 지도학습
: 모델을 가르치는 머신러닝 방식. 입력과 출력의 연관성을 찾게하는 학습. 
e.g) 이미지/음성 인식, 기계 번역

* 비지도학습
: 정답이 없는 데이터로 학습. 지도학습을 보조하는 용도
e.g) 군집 분석, 데이터 표현, 차원 감소

**지도학습과 비지도학습은 모두 정적인 데이터로 학습을 진행**

* 강화학습
: 보상과 처벌 체계를 세워 모델이 환경을 경험하며 배운다는 점에서 동적이다. 따라서, 풀고자하는 문제의 보상 방식을 잘 설계해야한다.

* 기타 학습방법
    * 준지도학습: 레이블이 부족한 데이터를 학습 
    * 메타학습: 학습법 자체를 연구
    * 원샷학습: 반복 없이 데이터를 한 번 혹은 몇 번만 보고 학습하는 원샷 학습
    * 전이학습: 문제를 위해 학습한 후 다른 문제에도 바로 적용하는 학습
    
# 1.3. 딥러닝과 신경망

**딥러닝 알고리즘 ⊂ 머신러닝 알고리즘**

정제되지 않은 데이터를 머신러닝 알고리즘을 입력하기 위해 데이터의 특징을 추출할 필요가 있다. 그러나 특징을 잘못 추출하는 경우 데이터의 중요한 특성을 잃어버릴 수 있기 때문에 경험과 노하우가 중요하다.

**=> "Feature Extraction"**

e.g) 이미지의 윤곽선 검출

* 인공신경망
: 사람의 뇌와 이를 구성하는 신경세포의 동작 원리에서 아이디어를 착안함. 인공신경망도 간단한 수식을 겹칩으로써 복잡한 일을 처리할 수 있게 된다.

## 1.3.1. 딥러닝의 제약사항

1. 데이터의 양
: 일반적인 알고리즘의 높은 정확도를 위해 엄청난 양의 학습데이터가 필요.

2. 시간과 하드웨어
: 폭발적으로 증가하는 데이터와 하드웨어의 성능 증가로 인해 딥러닝의 성능이 높아지게 된 계기

# 1.4. 파이토치의 개발

딥러닝 알고리즘은 기존에 계산되던 알고리즘과 달리 엄청난 양의 행렬에 대해 계산을 수행함. 이를 위해 **테아노(Theano)**라는 딥러닝 프레임워크를 발표했으며, 이는 keras와 같은 오픈소스 라이브러리들의 탄생에 기반을 제공함. 

파이토치는 위의 과정을 거쳐 텐서플로 이후에 탄생한 새로운 프레임워크이다.

# 1.5. 왜 파이토치인가?

파이토치는 파이썬 다운 특성을 잘 살린 딥러닝 프레임워크이다.

1. 파이토치의 철학은 **"가독성"**

2. 정적 / 동적 계산 그래프
    * 정적계산 그래프
    : 그래프 계산 방식을 최초에 정해둠으로써 최적화 하기 쉽다는 장점. 그러나, 텐서플로를 비롯한 대부분의 딥러닝 프레임워크는 파이썬을 사용하지만, 변수들이 지나가는 길인 그래프를 미리 정해두고 실행중에는 바꿀수 없어 C와 같은 느낌을 준다. 

    * 동적계산 그래프
    : 실행중에 그래프를 바꿀 수 있으며, 데이터에 유연한 모델을 만들 수 있다는 장점.
